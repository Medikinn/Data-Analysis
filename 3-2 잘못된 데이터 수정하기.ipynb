## 잘못된 값 바꾸기

```python
# 아래 코드는 오류 발생

data.astype({'발행년도': 'int32'})
```

```python
# '1988'이 포함된 행 수

data['발행년도'].str.contains('1988').sum()
```

```python
# '1988'만 포함된 행 수

sum(data['발행년도']=='1988')
```

```python
# 발행년도 열에서 숫자가 아닌 문자를 포함하는 모든 행

index = data['발행년도'].str.contains('\D', na=True)
print(index.sum()) # 숫자 이외의 문자가 들어간 행의 개수를 출력
data[index].head()
```

```python
# 발행년도 앞뒤에 있는 문자 제거

data2 = data.replace({'발행년도':r'.*(\d{4}).*'}, r'\1', regex=True)
data2[index].head()
```

```python
# 발행년도 열에서 숫자가 아닌 문자를 포함하는 모든 행

new_index = data2['발행년도'].str.contains('\D', na=True)
print(new_index.sum())
data2[new_index].head()
```

```python
data2.loc[new_index, '발행년도'] = '-1'
data2 = data2.astype({'발행년도': 'int64'})
```

```python
# 발행년도중 4000보다 큰 행 수 구하기 방법1

data2['발행년도'].gt(4000).sum()
```

```python
# 발행년도중 4000보다 큰 행 수 구하기 방법2

(data2['발행년도']>4000).sum()
```

```python
index4000 = data2['발행년도'].gt(4000)
data2.loc[index4000, '발행년도'] = data2.loc[index4000, '발행년도'] - 2333
```

```python
new_index400 = data2['발행년도'].gt(4000)
print(new_index400.sum())
data2[new_index400]
```

```python
data2.loc[new_index400, '발행년도'] = -1
```

```python
# 발행년도중 0보다 크고 1900보다 작은 큰 행 수 구하기 방법

index1900 = data2['발행년도'].gt(0) & data2['발행년도'].lt(1900)
data2[index1900]
```

```python
data2.loc[index1900, '발행년도'] = -1
```

```python
# 누락된 행 수

data2['발행년도'].eq(-1).sum()
```
