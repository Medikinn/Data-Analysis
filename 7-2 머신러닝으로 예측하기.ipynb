# 07-2 머신러닝으로 예측하기

## 모델 훈련하기

```python
import gdown

gdown.download('https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)
```

```python
import pandas as pd

data = pd.read_csv('ns_book7.csv', low_memory=False)
data.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(data, random_state=0)
```

```python
print(len(train), len(test))
```

```python
train1 = train[['도서권수']]
train2 = train['대출건수']

print(train1.shape, train2.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(train1, train2)
```

## 훈련된 모델을 평가하기: 결정계수

```python
test1 = test[['도서권수']]
test2 = test['대출건수']

lr.score(test1, test2)
```

```python
lr.fit(train2.to_frame(), train2)
lr.score(test2.to_frame(), test2)
```

## 연속적인 값 예측하기: 선형 회귀

```python
print(lr.coef_, lr.intercept_)
```

## 카테고리 예측하기: 로지스틱 회귀

```python
m = data['대출건수'].mean()
mtrain = train2 > m
mtest = test2 > m
```

```python
from sklearn.linear_model import LogisticRegression

lor = LogisticRegression()

lor.fit(train1, mtrain)
lor.score(test1, mtest)
```

```python
mtest.value_counts()
```

```python
from sklearn.dummy import DummyClassifier

dc = DummyClassifier()

dc.fit(train1, mtrain)
dc.score(test1, mtest)
```

## 평균제곱오차와 평균절댓값오차로 모델 평가하기

```python
lr.fit(train1, train2)
```

```python
pred = lr.predict(test1)
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(test2, pred)
```
