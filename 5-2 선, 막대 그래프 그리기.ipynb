## 연도별 발행 도서 개수 구하기

```python
import gdown

gdown.download('https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)
```

```python
import pandas as pd

data = pd.read_csv('ns_book7.csv', low_memory=False)
data.head()
```

```python
count_year=data['발행년도'].value_counts()
count_year
```

```python
import numpy as np

np.unique(data['발행년도'],return_counts=True)
```

```python
# 인덱스를 기준으로 오름차순

count_year=count_year.sort_index()
count_year
```

```python
# 발행년도가 2030 이하인 행 추출 방법1

count_year=count_year[count_year.index<=2030]
count_year
```

## 주제별 도서 개수 구하기

```python
import numpy as np

def first_char(i):
    if i is np.nan:
        return '-1'
    else:
        return i[0]

count_subject=data['주제분류번호'].apply(first_char).value_counts()
count_subject
```

## 막대 그래프 그리기

```python
import matplotlib.pyplot as plt

plt.bar(count_subject.index,count_subject.values)
plt.title('Books by subject')
plt.xlabel('subject')
plt.ylabel('number of books')

for idx, val in count_subject.items():
    plt.annotate(val, (idx, val), xytext=(0, 2), textcoords='offset points')
plt.show()
```

```python
# 막대 그래프 수정 (그래프 너비,색 변경),(데이터 레이블 크기,색,위치 변경)

import matplotlib.pyplot as plt

plt.bar(count_subject.index,count_subject.values,width=0.7, color='blue')
plt.title('Books by subject')
plt.xlabel('subject')
plt.ylabel('number of books')

for idx, val in count_subject.items():
    plt.annotate(val, (idx, val), xytext=(0, 2), textcoords='offset points',
                 fontsize=8, ha='center', color='green')
plt.show()
```

```python
# 가로 막대 그래프 (그래프 너비,색 변경),(데이터 레이블 크기,색,위치 변경)

import matplotlib.pyplot as plt

plt.barh(count_subject.index,count_subject.values, height=0.7, color='blue')
plt.title('Books by subject')
plt.xlabel('subject')
plt.ylabel('number of books')

for idx, val in count_subject.items():
    plt.annotate(val, (val, idx), xytext=(2, 0), textcoords='offset points',
                 fontsize=8, va='center', color='green')
plt.show()
```
