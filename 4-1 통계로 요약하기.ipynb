## 평균

$평균 = \dfrac{a + b + c}{3}$

$평균 = \dfrac{x_1 + x_2 + x_3}{3}$

```python
x = [10,20,30]
sum = 0

for i in range(3):
  sum += x[i]
print("평균:", sum / len(x))
```

$평균 = \dfrac{x_1 + x_2 + x_3}{3} = \dfrac{\sum_{i=1}^{3} x_i}{3}$

## 분산

$ s^2 = \dfrac{\sum_{i=1}^{n}(x_i - \bar{x})^2}{n}$

$ \bar{x} = \dfrac{\sum_{i=1}^n x_i}{n}$

## 표준 편차

$ s = \sqrt{\dfrac{\sum_{i=1}^{n}(x_i - \bar{x})^2}{n}}$

$\sqrt{4}=2$

```python
import numpy as np

diff = data1['대출건수'] - data1['대출건수'].mean()

np.sqrt( np.sum(diff**2) / (len(data1)-1) )
```

## 넘파이의 기술통계 함수

### 평균 구하기

```python
import numpy as np

np.mean(data1['대출건수'])
```

$\dfrac{국어점수 \times 2 + 수학점수}{3}$

$\dfrac{국어점수 \times 국어가중치 + 수학점수 \times 수학가중치}{국어가중치 + 수학가중치}$

$가중평균 = \dfrac{x_1 \times w_1 + x_2 \times w_2}{w_1 + w_2} = \dfrac{\sum_{i=1}^{2} x_i \times w_i}{\sum_{i=1}^{2} w_i}$

```python
np.average(data1['대출건수'], weights=1/data1['도서권수'])
```

```python
np.mean(data1['대출건수']/data1['도서권수'])
```

```python
data1['대출건수'].sum()/data1['도서권수'].sum()
```
