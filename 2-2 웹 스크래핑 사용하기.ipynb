## 검색 결과 페이지 가져오기

```python
#컬럼 추출 방법3

data.loc[0:1, 'bookname':'authors']
```

```python
#컬럼 추출 방법4

data.loc[:, 'no':'isbn13'].head()
```

```python
#컬럼 추출 방법5

data.loc[::2, 'no':'isbn13'].head()
```

## `merge()` 함수의 매개변수

```python
import pandas as pd

df1 = pd.DataFrame({'col1': ['a','b','c'], 'col2': [1,2,3]})
df1
```

```python
import pandas as pd

df2 = pd.DataFrame({'col1': ['a','b','d'], 'col3': [10,20,30]})
df2
```

**merge(df1, df2, on='공통열', how='방식')**

```python
pd.merge(df1, df2, on='col1')
```

**how='방식' → 병합 방식 (inner, left, right, outer)**

```python
pd.merge(df1, df2, on='col1', how='left')
```

```python
pd.merge(df1, df2, on='col1', how='right')
```

```python
pd.merge(df1, df2, on='col1', how='outer')
```

```python
pd.merge(df1, df2, left_on='col1', right_on='col1')
```

```python
pd.merge(df1, df2, left_on='col2', right_index=True)
```
