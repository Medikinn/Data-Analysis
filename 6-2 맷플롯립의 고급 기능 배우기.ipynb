## 하나의 피겨에 여러 개의 선 그래프 그리기

```python
top30 = data['출판사'].value_counts()[:30]
top30_idx = data['출판사'].isin(top30.index)
```

```python
# 컬럼 추출 및 그룹화

data1 = data[top30_idx][['출판사', '발행년도', '대출건수']]
data1 = data1.groupby(by=['출판사', '발행년도']).sum()
```

```python
# 인덱스 초기화
data1 = data1.reset_index()
data1[data1['출판사'] == '황금가지'].head()
```

```python
line1=data1[data1['출판사']=='황금가지']
line2=data1[data1['출판사']=='비룡소']
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(line1['발행년도'], line1['대출건수'], label='황금가지')
ax.plot(line2['발행년도'], line2['대출건수'], label='비룡소')
ax.set_title('년도별 대출건수')
fig.show()
```

```python
# 범례추가
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(line1['발행년도'], line1['대출건수'], label='황금가지')
ax.plot(line2['발행년도'], line2['대출건수'], label='비룡소')

ax.set_title('년도별 대출건수')

ax.legend()

fig.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

for i in top30.index[:5]:
    line = data1[data1['출판사']==i]
    ax.plot(line['발행년도'],line['대출건수'], label=i)

ax.set_title('년도별 대출건수')

ax.legend()

ax.set_xlim(1985,2025)

fig.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

for i in top30.index[:5]:
    line = data1[data1['출판사']==i]
    ax.plot(line['발행년도'],line['대출건수'], label=i)

ax.set_title('년도별 대출건수')

ax.legend()

ax.axis([1985,2025,-500,14000])

fig.show()
```

```python
# 특정 열의 특정 값 가져오기

top10 = top30.index[:10]
year_cols = data2.columns.get_level_values(1) # 년도 값 가져오기
```

```python
# 누적 영역 그래프 그리기

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.stackplot(year_cols, data2.loc[top10].fillna(0), labels=top10)

ax.set_title('년도별 대출건수')

ax.legend(loc='upper left')

ax.set_xlim(1985, 2025)

fig.show()
```

## 하나의 피겨에 여러 개의 막대 그래프 그리기

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.bar(line1['발행년도'], line1['대출건수'], label='황금가지')
ax.bar(line2['발행년도'], line2['대출건수'], label='비룡소')

ax.set_title('년도별 대출건수')

ax.legend()

fig.show()
```

```python
# 막대를 나란히 그리기

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.bar(line1['발행년도']-0.2, line1['대출건수'], width=0.4, label='황금가지')
ax.bar(line2['발행년도']+0.2, line2['대출건수'], width=0.4, label='비룡소')

ax.set_title('년도별 대출건수')

ax.legend()

fig.show()
```

```python
# 누적 막대 그래프 그리기 방법 1

height1 = [5, 4, 7, 9, 8]
height2 = [3, 2, 4, 1, 2]

plt.bar(range(5), height1, width=0.5)
plt.bar(range(5), height2, bottom=height1, width=0.5)
plt.show()
```

```python
# 누적 막대 그래프 그리기 방법 2

height3 = [a + b for a, b in zip(height1, height2)]

plt.bar(range(5), height3, width=0.5)
plt.bar(range(5), height1, width=0.5)
plt.show()
```

### 데이터값 누적하여 그리기

```python
# 상위 5개의 출판사의 2013 ~ 2020년 대출건수

data2.loc[top10[:5], ('대출건수',2013):('대출건수',2020)]
```

```python
# 연도별 누적 대출 건수 (열 누적)

data2.loc[top10[:5], ('대출건수',2013):('대출건수',2020)].cumsum()
```

```python
# 출판사 누적 대출 건수 (행 누적)

data2.loc[top10[:5], ('대출건수',2013):('대출건수',2020)].cumsum(axis=1)
```

```python
data3 = data2.loc[top10].cumsum()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

for i in reversed(range(len(data3))): # reversed: 순서 역전
    bar = data3.iloc[i]     # 행 추출
    label = data3.index[i]  # 출판사 이름 추출
    ax.bar(year_cols, bar, label=label)

ax.set_title('년도별 대출건수')

ax.legend(loc='upper left') # 범례위치 조정

ax.set_xlim(1990, 2025)

fig.show()
```
